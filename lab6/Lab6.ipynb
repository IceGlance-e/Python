```python
import pandas as pd
import numpy as np 
import re
import pandas as pd
```

```python
#Завдання 1
books = pd.read_csv("books.csv")
books
```

```python
#Завдання 2
books['Title_c']=pd.Series(list(books['Title'].agg({'Title': lambda x: len(re.findall(r'[a,e,i,o,u,y]', x.lower()))})), index=books.index)
books[books['Title_c']<=1].drop(books.columns[6:10], axis=1)
```

```python
#Завдання 3
books['Author_c2']=pd.Series(list(books['Author'].str.split().str[0].str[0].str.findall(r'[A,E,I,O,U,Y]').str.len()+books['Author'].str.split().str[0].str[-2].str.findall(r'[a,e,i,o,u,y]').str.len()), index=books.index)
books[books['Author_c2']==0].drop(books.columns[6:10], axis=1)
```

```python
#Завдання 4
books[books['Title_c']==books['Title_c'].max()].drop(books.columns[6:10], axis=1)
```

```python
#Завдання 5
books[books['Title'].str.len()-books['Title'].str.count(" ")==3].drop(books.columns[6:10], axis=1)
```

```python
#Завдання 6
books['A_3']=pd.Series(list(books['Author'].str.lower().str.split().str[-1].str.findall(r'[bcdfghjklmnpqrstvwxz]{3}').str.len()), index=books.index)
books[books['A_3']==1].drop(books.columns[6:10], axis=1)
```

```python
#Завдання 7
books['Author_c']=pd.Series(list(books['Author'].agg({'Author': lambda x: x.str.split().str[-1]})), index=books.index)
books.drop(books.columns[[6,7,9]], axis=1)
```

```python
#Завдання 8
books[books['Author'].str.split().str[0].str.len()==6].drop(books.columns[6:10], axis=1)
```
