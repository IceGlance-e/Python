```python
import pandas as pd
import numpy as np 
```

```python
#Задание 1
df1 = pd.DataFrame({
'Name': ['Ivan Ivanov', 'Petro Petrov', 'Katerina Kulbaba', 'Pavlo Sydorenko', 'Ivan Dubenko', 'Makar Khmara'],
'Year': [2018, 2018, 2019, 2018, 2019, 2020],
'Country': ['Egypt', 'Egypt', 'Egypt', 'Turkey', 'Turkey', 'Turkey'],
'Agency': ['Sun', 'Moon', 'Star', 'Sun', 'Star', 'Star']
})
df1
```

```python
#Задание 1
df2 = pd.DataFrame({
'Country': ['Egypt', 'Egypt', 'Turkey', 'Turkey', 'Ukraine', 'Ukraine'],
    'City': ['Cairo', 'Luxor', 'Istanbul', 'Izmir', 'Lviv', 'Kharkiv'],
'Days to stay': [5, 4, 5, 3, 4, 3],
    'Transport': ['Train', 'Bus', 'Airplane', 'Airplane', 'Boat', 'Car'],
    'Tour cost': [123423, 234234, 456454, 734533, 734534, 644522]
})
df2
```

```python
#Задание 2
result = pd.merge(df1, df2 , how='outer', on='Country')
result.set_index('Country')
```

```python
#Задание 3
result=result.ffill(axis = 0)
result
```

```python
#Задание 4
tuple(result.groupby('Country')['Tour cost'].mean().sort_values().items())[0]
```

```python
#Задание 5
result['Cost per day']=result['Tour cost']/result['Days to stay']
result
```

```python
#Задание 6
tuple(result[['Name','Country','Cost per day']].sort_values(by=['Cost per day'], ascending=False).reset_index(drop=True).loc[1])
```

```python
#Задание 7
result[result['City']=="Izmir"].reset_index(drop=True).shape[0]
```
